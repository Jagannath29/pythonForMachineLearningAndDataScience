```python
# y = 2x + 3

# when x = 1

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(1,6)

y_hat = 2 * x + 3

df = pd.DataFrame(x, y_hat,)
df
```

```python
plt.plot(x, y_hat, marker='o')
plt.ylabel('Y value')
plt.xlabel('X value')
plt.title('Simple Linear Regression')
plt.show()
```
