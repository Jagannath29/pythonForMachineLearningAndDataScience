```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Question:

<li>Read cricket_scores.csv and load it into a pandas dataframe.</li>
<li>Extract the scores and overs details when Newzeland was batting.</li>
<li>Create a lineplot to show the scores scored by Newzeland in each over.</li>
<li>In the x-axis, we have to include the overs while in y-axis, you have to include the scores.</li> 

```python
cricket_df = pd.read_csv('cricket_scores.csv')
cricket_df
```

```python
newzerland_detail = cricket_df.loc[cricket_df['Batting_team'] == 'Newzeland']
newzerland_detail
```

```python
newzerland_detail.shape
```

```python
nz_runs = newzerland_detail['Run']
nx_over = newzerland_detail['Over']
```

```python
plt.plot(nx_over, nz_runs, marker = 'x', markersize = 10)
plt.title('Run in each over')
plt.xlabel('Over')
plt.ylabel('Runs')
plt.show()
```

The runrate is increasing linearly.
