```python
# 2x^2 + 3x^3 + 4x + 5

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
 
```

```python
x = np.arange(1,6)
y = 2*x **2 + 3*x**3 + 4*x + 5
```

```python
# Create scatterplot
plt.plot(x,y, color='red', marker= 'o' )
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Multiple Polynomial Regression")
plt.show()
```
