```python
# a1x^3+a2x^2+a3x

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
 
```

```python


# Generate x and y
x = np.arange(1,6)
y = x ** 4 + x**3+x**2+ 3*x



# Create scatterplot
plt.plot(x,y, color='red', marker= 'o' )
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Polynomial Regression")
plt.show()
```
